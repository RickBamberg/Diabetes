```python
# Lista das colunas onde zero  implausvel e precisa ser tratado
colunas_para_imputar = ['Glicose', 'Presso arterial', 'Espessura da pele', 'Insulina', 'IMC']
print("Valores mdios ANTES da imputao (incluindo zeros onde aplicvel):")
print(df[colunas_para_imputar].mean()) # Apenas para comparao depois

print("\nAntes - estatsticas descritivas (colunas selecionadas):")
print(df[colunas_para_imputar].describe())

print("-" * 30)

for coluna in colunas_para_imputar:
    # 1. Seleciona a coluna
    series_coluna = df[coluna]

    # 2. Calcula a mediana IGNORANDO os zeros
    #    (Criamos uma cpia temporria sem zeros para calcular a mediana)
    mediana_coluna = series_coluna[series_coluna > 0].median()

    # Verificao se a mediana foi calculada (caso todos os valores > 0 fossem NaN ou a coluna s tivesse 0)
    if pd.isna(mediana_coluna):
        # Se no puder calcular a mediana (ex: s tem zeros),
        # voc pode usar a mdia geral (menos ideal) ou outra estratgia.
        # Mas para estas colunas,  quase certo que haver valores > 0.
        print(f"Aviso: No foi possvel calcular a mediana no-zero para '{coluna}'. Pulando imputao.")
        continue # Pula para a prxima coluna

    print(f"Coluna: '{coluna}' - Mediana (sem zeros): {mediana_coluna:.2f}")

    # 3. Faz o replace dos zeros com a mediana calculada para ESSA coluna
    #    Usamos .loc para evitar SettingWithCopyWarning e garantir a modificao no df original
    df.loc[df[coluna] == 0, coluna] = mediana_coluna

    # Alternativa (funciona tambm, mas .loc  preferido):
    # df[coluna] = df[coluna].replace(0, mediana_coluna)

print("-" * 30)
print("Imputao concluda.")

# 4. Verificao Opcional: Contar zeros restantes e ver novas estatsticas
print("\nZeros restantes nas colunas imputadas:")
print((df[colunas_para_imputar] == 0).sum()) # Deve ser tudo zero

print("\nValores mdios DEPOIS da imputao:")
print(df[colunas_para_imputar].mean()) # Compare com antes

print("\nNovas estatsticas descritivas (colunas selecionadas):")
print(df[colunas_para_imputar].describe())
```

## Separao dos dados

```python
# DataFrame Original (sem padronizao)
df_original = df.copy()

# Separao do Target dos Features
X = df.drop('Resultado', axis=1) # Features
Y = df['Resultado']              # Target
print('Separao do Target dos Features')
```

```python
# Separa TODO o conjunto (X e Y juntos, mas mantendo a correspondncia)
# em 70% para treino e 30% para teste.
X_train_df, X_test_df, Y_train_df, Y_test_df = train_test_split(
    X, Y,
    test_size=0.3,   # 30% dos dados vo para o conjunto de teste
    random_state=42, # Para reprodutibilidade do embaralhamento
    stratify=Y       # IMPORTANTE para classificao: mantm a proporo das classes em treino/teste
)
print('Separao do Treino e Teste 70/30')
```

```python
#from matplotlib import pyplot as plt
plt.figure(figsize=(16,6))
ax = sns.boxplot(data = df_original)
```

```python
# Padrozinao StandardScaler
scaler_std = StandardScaler()
scaler_std.fit(X_train_df)
X_train_df_pad = scaler_std.transform(X_train_df)
X_test_df_pad = scaler_std.transform(X_test_df)
print('Transformao do DataFrame Padronizado')
```

```python
# Balanceamento do X_train_df_padrao -- SMOTE
smote = SMOTE(random_state=42)
X_train_df_pad_bal, Y_train_df_pad_bal = smote.fit_resample(X_train_df_pad, Y_train_df)
print("\nDistribuio no Treino Aps SMOTE:")
print(pd.Series(Y_train_df_pad_bal).value_counts())
print('Balanceamento do DataFrame Padronizado')
```

```python
# Padronizao Mista
# Defina as colunas para cada tipo de transformao (usando os nomes ORIGINAIS)
colunas_robust = ['Insulina', 'Espessura da pele', 'IMC']
colunas_standard = ['Glicose', 'Presso arterial']
colunas_passthrough = ['Gravidez', 'Idade']

# Aplique as transformaes SEM renomear ainda
transformer = ColumnTransformer(
    transformers=[
        ('robust', RobustScaler(), colunas_robust),
        ('standard', StandardScaler(), colunas_standard),
        ('pass', 'passthrough', colunas_passthrough)
    ],
    remainder='passthrough'
)

transformer.fit(X_train_df)

X_train_df_misto = transformer.transform(X_train_df)
X_test_df_misto = transformer.transform(X_test_df)
print('Transformao do DataFrame Misto')
```

```python
# Balanceamento do X_train_df_misto -- SMOTE
smote = SMOTE(random_state=42)
X_train_df_misto_bal, Y_train_df_misto_bal = smote.fit_resample(X_train_df_misto, Y_train_df)
print("\nDistribuio no Treino Aps SMOTE:")
print(pd.Series(Y_train_df_misto_bal).value_counts())
print('Balanceamento do DataFrame Misto')
```

```python
# Monta DataFrame df_misto
df_misto = pd.DataFrame(transformer.transform(X), columns=transformer.get_feature_names_out())
df_misto['Resultado'] = Y.values
df_misto.head()
```
